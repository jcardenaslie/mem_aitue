```python
#get paginations
def get_page_links(soup):
    paginations = soup.find('ul',class_='pagination').find_all('a')
    pages_links = []
    for anchor in paginations:
        pages_links.append(anchor.get('href'))
    return pages_links
```

```python
def get_page_products_links(page):
    products_items = soup.find_all('div',class_='product-item')
    len(products_items)
    products_links = []
    for item in products_items:
        item_data = item.find('div', class_='product-item-summary')
        products_links.append(primary_url + item_data.h4.a.get('href'))
    return products_links
```

```python
page_links = get_page_links(soup)
print(page_links)
products_links = []

for link in page_links:
#     print(primary_url + link)
    proxies_req = Request(primary_url + link)
    proxies_doc = urlopen(proxies_req).read().decode('utf8')
    soup = BeautifulSoup(proxies_doc, 'html.parser')
    products_links.extend(get_page_products_links(soup))
```

```python
def get_building_data(link):
    proxies_req = Request(link)
    proxies_doc = urlopen(proxies_req).read().decode('utf8')
    soup = BeautifulSoup(proxies_doc, 'html.parser')
    
    name = soup.find('h1',attrs={'role':'heading'}).text
    name
    code = soup.find('span',class_='prj-code').text
    code
    from_price = soup.find('span', class_='prj-price-range-lower').text
    from_price
    building_features = soup.find_all('span',class_='project-feature-details')
    building_features
    address = soup.find_all('p',class_='prj-map-addr-obj')
    address
    description = soup.find('div',class_='prj-highlights-panel').find('p')
    description #separar
    other_info = soup.find_all('div',class_='prj-other-info-item')
    other_info # separar
    data = [name,code,from_price,building_features,address,description,other_info]
    return data
```
